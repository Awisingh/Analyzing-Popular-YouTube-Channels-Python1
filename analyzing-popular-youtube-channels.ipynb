# import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/most-subscribed-youtube-channel/Most Subscribed YouTube Channels_exported.csv')
```

# data visualization

```python
plt.figure(figsize=(12, 6))  
sns.countplot(data=df, x='Category', palette='Set1')
plt.xticks(rotation=40)
plt.title('best category of you Tube')
```

# graph indicates that the 'Music' category is the most popular among YouTube channels."

```python
plt.figure(figsize=(12, 6))  
top_countries = df['Country'].value_counts().head(8).index
df_top_countries = df[df['Country'].isin(top_countries)]
sns.countplot(data=df_top_countries, x='Country', palette='Set1')
plt.xticks(rotation=45)
plt.title('best category of you Tube')
```

# The majority of the top YouTube channels are from India."**

```python
top_10_channels = df.sort_values(by='Subscribers (millions)', ascending=False).head(10)
colors = plt.cm.tab10(np.arange(len(top_10_channels)))

plt.figure(figsize=(12, 6))  
plt.bar(top_10_channels['Name'], top_10_channels['Subscribers (millions)'], color=colors)
plt.xticks(rotation=90)
```

```python
highest_subscribed_per_category = df.loc[df.groupby('Category')['Subscribers (millions)'].idxmax()]
print(highest_subscribed_per_category[['Category', 'Name', 'Subscribers (millions)']])
```

# best youtube chaanel category wise

```python
highest_subscribed_per_category = df.loc[df.groupby('Category')['Subscribers (millions)'].idxmax()]
plt.figure(figsize=(12, 6))
sns.barplot(data=highest_subscribed_per_category, x='Category', y='Subscribers (millions)', hue='Name', dodge=False)
plt.xticks(rotation=45)
plt.title('Highest Subscribed Channel in Each Category')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
